```python
import pandas as pd

df = pd.read_csv('D:\\DATA\\pyth_nayak\\1-Python Basics\\data_services.csv',header = None)
df
```

```python
t = df[0]
v = df[1]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12,8))
c = ['g','orange','y','m','b','r','c','brown','black']
plt.bar(t,v, width = 0.4, color=c)
plt.xlabel('SEVICES')
plt.ylabel('QTY (IN 1000s)')
plt.title('Till YEAR 2024')

plt.xticks(fontsize=10)


plt.show()
```
